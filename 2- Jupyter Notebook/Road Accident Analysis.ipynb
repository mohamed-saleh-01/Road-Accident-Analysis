Outliers Values

```python
def detect_outliers_iqr (df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[columns] < lower) | (df[columns] > upper)]
    return outliers

outliers_speed = detect_outliers_iqr(df,'Vehicle_Speed')
print(len(outliers_speed))
outliers_limit = detect_outliers_iqr(df,'Speed_Limit')
print(len(outliers_limit))
outliers_emerg = detect_outliers_iqr(df,'Time_Taken_for_Emergency_Response')
print(len(outliers_emerg))
outliers_vic = detect_outliers_iqr(df,'Num_Vehicles_Involved')
print(len(outliers_vic))
outliers_casu = detect_outliers_iqr(df,'Num_Casualties')
print(len(outliers_casu))
```

```python
def remove_outliers_iqr (df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[columns] < lower) & (df[columns] > upper)]

# df = remove_outliers_iqr(df,'Vehicle_Speed')
```
